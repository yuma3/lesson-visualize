```python
messi = pd.DataFrame(messi90)
messi['x'] = messi['x']*1.2
messi['y'] = messi['y']*.8
messi['endX'] = messi['endX']*1.2
messi['endY'] = messi['endY']*.8
messi
```

```python
fig, axes = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('black')
axes.patch.set_facecolor('black')

pitch = Pitch(
    pitch_type='statsbomb', 
    orientation='horizontal',
    pitch_color='black',
    line_color='white',
    figsize=(13,8),
    constrained_layout=False, 
    tight_layout=True
)
pitch.draw(ax=axes)
plt.gca().invert_yaxis()


for x in range(len(messi['x'])):
    if messi['outcome'][x]=='s':
        plt.plot((messi['x'][x], messi['endX'][x]),(messi['y'][x], messi['endY'][x]), color='dodgerblue')
        plt.scatter(messi['x'][x], messi['y'][x], color='dodgerblue')
    if messi['outcome'][x]=='u':
        plt.plot((messi['x'][x], messi['endX'][x]), (messi['y'][x],messi['endY'][x]),color='tomato')
        plt.scatter(messi['x'][x], messi['y'][x], color='tomato')
        
plt.title('Juventus vs Barcelona [Pass Map]', color='white', size=20)  
plt.text(1,8, 'Juventus vs Barcelona in CL2020' , color='white', fontsize=20, alpha=0.9)
plt.text(96,1, '@Bucciaratimes', color='white',fontsize=14) 
plt.text(1,1, 'Lionel Messi [Pass Map]', color='white',fontsize=18)

fig.savefig('/work/output/messi_passmap90.png')
```

```python
fig, ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('black')
ax.patch.set_facecolor('black')

pitch = Pitch(
    pitch_type='statsbomb', 
    orientation='horizontal',
    pitch_color='black',
    line_color='white',
    figsize=(13,8),
    constrained_layout=False, 
    tight_layout=True
)
pitch.draw(ax=ax)

# kde = sns.kdeplot(
#     df['x'],
#     df['y'],
#     shade=True,
#     shade_lowest=False,
#     alpha=0.5,
#     n_levels=10,
#     cmap='rainbow'
# )

for i in range(len(df['x'])):
    if i == 3:
        plt.plot((df['x'][3], df['dx'][3]),(df['y'][3], df['dy'][3]), color='tomato')
        plt.scatter(df['x'][0], df['y'][0], color='tomato')
    if df['outcome'][i]=='success':
        plt.plot((df['x'][i], df['dx'][i]),(df['y'][i], df['dy'][i]), color='dodgerblue')
        plt.scatter(df['x'][i], df['y'][i], color='tomato')
    if df['outcome'][i]=='unsuccess':
        plt.plot((df['x'][i], df['dx'][i]),(df['y'][i], df['dy'][i]), color='tomato')
        plt.scatter(df['x'][i], df['y'][i], color='tomato')

plt.xlim(0,120)
plt.ylim(0,80)
        
plt.title('Griezmann vs Osasuna [Pass Map]', alpha=0.8, color='white', size=22)  
plt.text(1,8, 'Griezmann vs Osasuna [Pass Map]', alpha=0.9, color='white', fontsize=20)
plt.text(1,1, '@Bucciaratimes', color='white',fontsize=14, alpha=0.9) 

  
```
