```python
fig, axes = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('black')
axes.patch.set_facecolor('black')

pitch = Pitch(
    pitch_type='statsbomb', 
    orientation='horizontal',
    pitch_color='black',
    line_color='white',
    figsize=(13,8),
    constrained_layout=False, 
    tight_layout=True
)
pitch.draw(ax=axes)
plt.gca().invert_yaxis()


for x in range(len(messi['x'])):
    if messi['outcome'][x]=='s':
        plt.plot((messi['x'][x], messi['endX'][x]),(messi['y'][x], messi['endY'][x]), color='dodgerblue')
        plt.scatter(messi['x'][x], messi['y'][x], color='dodgerblue')
    if messi['outcome'][x]=='u':
        plt.plot((messi['x'][x], messi['endX'][x]), (messi['y'][x],messi['endY'][x]),color='tomato')
        plt.scatter(messi['x'][x], messi['y'][x], color='tomato')
        
plt.title('Juventus vs Barcelona [Pass Map]', color='white', size=20)  
plt.text(1,8, 'Juventus vs Barcelona in CL2020' , color='white', fontsize=20, alpha=0.9)
plt.text(96,1, '@Bucciaratimes', color='white',fontsize=14) 
plt.text(1,1, 'Lionel Messi [Pass Map]', color='white',fontsize=18)

fig.savefig('/work/output/messi_passmap90.png')
```
